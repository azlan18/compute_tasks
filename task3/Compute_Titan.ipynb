# **Data Visualization with Matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
# Create a bar plot for Pclass count
pclass_counts = df["Pclass"].value_counts()
plt.figure(figsize=(8, 4))
plt.bar(pclass_counts.index, pclass_counts.values, color='blue')
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.title("Passenger Class Distribution")
plt.xticks(sorted(df["Pclass"].unique()))
plt.show()
```

```python
# Create a histogram for passenger ages
plt.figure(figsize=(8, 4))
plt.hist(df["Age"], bins=20, color='skyblue')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Passenger Age Distribution")
plt.show()
```

```python
# Count the number of male and female passengers using pie chart
gender_counts = df["Sex"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightpink', 'lightblue'])
plt.title("Distribution of Male and Female Passengers")
plt.show()
```

```python
# Create a box plot for the "Fare" column
plt.figure(figsize=(8, 6))
plt.boxplot(df["Fare"], vert=False, widths=0.7, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.xlabel("Fare")
plt.title("Fare Distribution (Box Plot)")
plt.show()
```

# **. Data Visualization with Seaborn:**

```python
import seaborn as sns
```

```python
df = pd.read_csv("test.csv")

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Create a violin plot for the distribution of ages by passenger class
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x="Pclass", y="Age", palette="Set2")
plt.title("Distribution of Ages by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
df3 = pd.read_csv("train.csv")

# Create a count plot to differentiate survivors and non-survivors by passenger class
plt.figure(figsize=(8, 6))
sns.countplot(data=df3, x="Survived", hue="Pclass", palette="Set1")
plt.title("Survivors vs. Non-Survivors by Passenger Class")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.xticks([0, 1], ["Non-Survivors", "Survivors"])
plt.legend(title="Passenger Class", loc="upper right", labels=["1st Class", "2nd Class", "3rd Class"])
plt.show()
```

```python
df = pd.read_csv("test.csv")

# Select numerical features (assuming all numeric columns are relevant)
numerical_features = df.select_dtypes(include=['float64', 'int64'])

# Create a pair plot
sns.set(style="ticks")
sns.pairplot(data=df, vars=numerical_features.columns)
plt.suptitle("Pair Plot of Numerical Features", y=1.02)
plt.show()
```
